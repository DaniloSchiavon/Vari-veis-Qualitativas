1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('G:\EBAC\Provas\Modulo 10\previsao_de_renda.csv')

df['data_ref'] = pd.to_datetime(df['data_ref'])

# Check de datas invalidas
datas_invalidas = df['data_ref'].apply(lambda x: pd.isnull(pd.to_datetime(x, errors='coerce')))

print(df[datas_invalidas])

# Converter as datas validas
df.loc[~datas_invalidas, 'data_ref'] = pd.to_datetime(df.loc[~datas_invalidas, 'data_ref'])

# Convertendo a coluna 
qualitativa_vars = ['posse_de_veiculo', 'posse_de_imovel', 'sexo', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']
````

````python
# Gráfico de barras p cada variável

df['posse_de_veiculo'] = df['posse_de_veiculo'].astype(str)
df['posse_de_imovel'] = df['posse_de_imovel'].astype(str)
df['sexo'] = df['sexo'].astype(str)
df['tipo_renda'] = df['tipo_renda'].astype(str)
df['educacao'] = df['educacao'].astype(str)
df['estado_civil'] = df['estado_civil'].astype(str)
df['tipo_residencia'] = df['tipo_residencia'].astype(str)

# Gráfico de barras p cada variável

def plot_countplot(variable):
  plt.figure(figsize=(14, 7))
  ax = sns.countplot(x='data_ref', hue=variable, data=df)
  tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
  ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
  labels = ax.set_xticklabels(tick_labs, rotation=45)
  plt.title(f'Distribuição de {variable} ao longo do tempo')
  plt.xlabel('Data de Referência')
  plt.ylabel('Contagem')
  plt.legend(title=variable, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
  plt.show()

# Plotando cada variável qualitativa
plot_countplot('posse_de_veiculo')
plot_countplot('posse_de_imovel')
plot_countplot('sexo')
plot_countplot('tipo_renda')
plot_countplot('educacao')
plot_countplot('estado_civil')
plot_countplot('tipo_residencia')
````

Observações:

### 1) Podemos notar que, as pessoas com nível de educação 'secundário' e 'superior completo' tiveram variações muito mais abruptas ao longo do tempo em relação aos demais grupos.
### 2) No tipo de residência, nos grupos 'comunitários' e 'estúdio' também tiveram variações maiores, proporcionalmente, em relação aos demais grupos observados.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
for var in qualitativa_vars:
    plt.figure(figsize=(14, 7))
    sns.lineplot(x='data_ref', y='renda', hue=var, data=df, ci=None)
    plt.title(f'Médias de Renda por {var} ao longo do tempo')
    plt.xlabel('Data de Referência')
    plt.ylabel('Renda Média')
    plt.legend(title=var)
    plt.xticks(rotation=45)
    plt.show()
````
